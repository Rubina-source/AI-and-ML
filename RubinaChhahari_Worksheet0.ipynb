```python
# 4 TO - DO - Task
# 4.1 Exercise on Functions:
# Task 1 – Unit Conversion Program

def unit_converter():
    """
    Converts between different units of measurement.

    The function allows the user to select a conversion type (length, weight, or volume)
    and input a value. It then performs the conversion and prints the result.

    Returns:
        float: Converted value or None if invalid input.
    """

    print("Select conversion type:")
    print("1. Length")
    print("2. Weight")
    print("3. Volume")

    choice = input("Enter choice (1/2/3): ")

    try:
        value = float(input("Enter value to convert: "))

        if choice == "1":
            print("1. Meters to Feet")
            print("2. Feet to Meters")
            option = input("Choose option: ")

            if option == "1":
                result = value * 3.28084
                print("Result:", result, "ft")
                return result

            elif option == "2":
                result = value / 3.28084
                print("Result:", result, "m")
                return result

        elif choice == "2":
            print("1. Kilograms to Pounds")
            print("2. Pounds to Kilograms")
            option = input("Choose option: ")

            if option == "1":
                result = value * 2.20462
                print("Result:", result, "lbs")
                return result

            elif option == "2":
                result = value / 2.20462
                print("Result:", result, "kg")
                return result

        elif choice == "3":
            print("1. Liters to Gallons")
            print("2. Gallons to Liters")
            option = input("Choose option: ")

            if option == "1":
                result = value * 0.264172
                print("Result:", result, "gal")
                return result

            elif option == "2":
                result = value / 0.264172
                print("Result:", result, "L")
                return result

        else:
            print("Invalid conversion type")

    except ValueError:
        print("Invalid input. Please enter numeric values.")
        return None
```

```python
#Task 2 – Mathematical Operations on List

def find_sum(numbers):
    """Returns the sum of numbers in the list."""
    return sum(numbers)

def find_average(numbers):
    """Returns the average of numbers in the list."""
    return sum(numbers) / len(numbers)

def find_max(numbers):
    """Returns the maximum value from the list."""
    return max(numbers)

def find_min(numbers):
    """Returns the minimum value from the list."""
    return min(numbers)


def math_operations():
    """
    Prompts the user to choose a mathematical operation and input numbers.
    Performs the selected operation and displays the result.
    """

    print("Choose operation:")
    print("1. Sum")
    print("2. Average")
    print("3. Maximum")
    print("4. Minimum")

    choice = input("Enter choice: ")

    try:
        numbers = list(map(float, input("Enter numbers separated by spaces: ").split()))

        if len(numbers) == 0:
            print("List cannot be empty")
            return

        if choice == "1":
            print("Sum =", find_sum(numbers))

        elif choice == "2":
            print("Average =", find_average(numbers))

        elif choice == "3":
            print("Maximum =", find_max(numbers))

        elif choice == "4":
            print("Minimum =", find_min(numbers))

        else:
            print("Invalid choice")

    except ValueError:
        print("Invalid input. Please enter numeric values.")
```

```python
# Exercise 4.2 – List Manipulation
def extract_every_other(lst):
    """Returns every other element starting from first."""
    return lst[::2]


def get_sublist(lst, start, end):
    """Returns sublist from start to end index inclusive."""
    return lst[start:end+1]


def reverse_list(lst):
    """Returns reversed list using slicing."""
    return lst[::-1]


def remove_first_last(lst):
    """Removes first and last element."""
    return lst[1:-1]


def get_first_n(lst, n):
    """Returns first n elements."""
    return lst[:n]


def get_last_n(lst, n):
    """Returns last n elements."""
    return lst[-n:]


def reverse_skip(lst):
    """Returns every second element from second last backwards."""
    return lst[-2::-2]
```

```python
# Exercise 4.3 – Nested List
def flatten(lst):
    """Flattens a nested list into a single list."""
    result = []
    for i in lst:
        if isinstance(i, list):
            result.extend(flatten(i))
        else:
            result.append(i)
    return result
```

```python
# Access Nested Element
def access_nested_element(lst, indices):
    """
    Returns element from nested list based on index path.
    """
    element = lst
    for i in indices:
        element = element[i]
    return element
```

```python
# Sum of Nested List
def sum_nested(lst):
    """Returns sum of all elements in nested list."""
    total = 0
    for i in lst:
        if isinstance(i, list):
            total += sum_nested(i)
        else:
            total += i
    return total
```

```python
# Remove Element from Nested List
def remove_element(lst, elem):
    """Removes all occurrences of elem from nested list."""
    result = []
    for i in lst:
        if isinstance(i, list):
            result.append(remove_element(i, elem))
        else:
            if i != elem:
                result.append(i)
    return result
```

```python
# Find Maximum in Nested List
def find_max(lst):
    """Finds maximum element in nested list."""
    flat = flatten(lst)
    return max(flat)
```

```python
# Count Occurrences
def count_occurrences(lst, elem):
    """Counts occurrences of element in nested list."""
    count = 0
    for i in lst:
        if isinstance(i, list):
            count += count_occurrences(i, elem)
        else:
            if i == elem:
                count += 1
    return count
```

```python
# Deep Flatten
def deep_flatten(lst):
    """Flattens deeply nested list."""
    result = []
    for i in lst:
        if isinstance(i, list):
            result.extend(deep_flatten(i))
        else:
            result.append(i)
    return result
```

```python
# Nested Average
def average_nested(lst):
    """Returns average of elements in nested list."""
    flat = deep_flatten(lst)
    return sum(flat) / len(flat)
```

# Solve Linear Equation System
# Equations:
2x − 3y + z = −1
x − y + 2z = −3
3x + y − z = 9

```python
A = np.array([
    [2,-3,1],
    [1,-1,2],
    [3,1,-1]
])

B = np.array([-1,-3,9])

# Solve AX = B
X = np.linalg.solve(A,B)

print("Solution [x,y,z] =", X)
```

```python
# Solve using Inverse Method
A_inv = np.linalg.inv(A)

X = np.dot(A_inv,B)

print("Solution using inverse =", X)
```
